```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# X = [63,66,69,71,71,75]
# Y = [127,142,162,156,169,208]
# X = [95,85,80,70,60]
# Y = [85,95,70,65,70]
X = [2005,2006,2007,2008,2009,2010]
Y = [12,19,29,37,45,49]
df = pd.DataFrame({
    'X':X,
    'Y':Y
})
df
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(df.X,df.Y)
```

```python
X = df.drop('Y',axis='columns')
X.head()
```

```python
Y = df.Y
Y.head()
```

```python
model = linear_model.LinearRegression()
model.fit(X,Y)
```

```python
print("y-intercept", model.intercept_)
print("x-intercept", model.coef_)
```

```python
model.predict([[2013]])
print(model.coef_*2013 + model.intercept_)
```

```python
prediction = model.predict(X)
prediction
```

```python
residual = Y - prediction
np.sum(residual**2)
```

```python
# Vectorised Linear Regression
# X = np.transpose(np.array([[1,1,1,1,1],[95,85,80,70,60]]))
# Y = np.array([85,95,70,65,70])
X = np.transpose(np.array([[1,1,1,1,1],[0,1,2,3,4]]))
Y = np.array([2,3,5,4,6])

Xt = np.transpose(X)
Xt_X = np.array(np.matmul(Xt,X))
Xt_X_inv = np.linalg.inv(Xt_X)
Xt_Y = np.matmul(Xt,Y)
theta = np.matmul(Xt_X_inv,Xt_Y)
theta
```

```python
def gradientDescent(X,Y,learningRate = 0.1,iter=1000):
    theta0 = 0
    theta1 = 0
    m = len(X)
    for i in range(iter):
        hypothesis = theta0 + theta1*X
        theta0 = theta0 -(-1/m) *learningRate*sum(Y-hypothesis)
        theta1 = theta1 - (-1/m) *learningRate*sum(X*(Y - hypothesis))
        print(theta0,theta1)

X = np.array([0,1,2,3,4])
Y = np.array([2,3,5,4,6])

X
Y
plt.scatter(X,Y)

gradientDescent(X,Y)
```
