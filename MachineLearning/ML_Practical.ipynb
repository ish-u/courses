```python
# We are intrested in finding a Linear Model that will detect the Price at which 
# the house will be sold at for a given Area
# According to the given data description in the data_description.txt we choose
#       Area  - GrLivAr Column - Independent Variable
#       Price - SalePrice - Dependent Variable
df_area_price = df[['GrLivArea','SalePrice']]
df_area_price = df_area_price.rename(columns={'GrLivArea':'area', 'SalePrice':'price'})
df_area_price
```

```python
# Missing/Incosistent and Special Values
# NA Values
(df_area_price.isna().sum())
```

```python
# Removing the Outlier 
df_area_price = df_area_price[df_area_price['area'] < 4500]
df_area_price = df_area_price[df_area_price['price'] < 600000]
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df_area_price['area'], df_area_price['price'])
```

```python
X_train,X_test,y_train,y_test = train_test_split(df_area_price.drop("price",axis=1),df_area_price.price,test_size=0.20)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train,y_train)
```

```python
print("Intercept and Slope from scikit library function")
print("b : ",model.intercept_,"\nm :", model.coef_)
plt.plot(X_train.to_numpy(), model.predict(X_train),color='r')
plt.scatter(X_train.to_numpy(), y_train,color='b')
```

##### Evaluation

```python
plt.plot(X_test.to_numpy(), model.predict(X_test),color='r')
plt.scatter(X_test.to_numpy(), y_test,color='b')
```

```python
model.score(X_test,y_test)
```

## Practical 15
Use some function for regularization of dataset based on problem 14. 


##### Preprocessing

```python
iris = datasets.load_iris()
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
renameDictionary = {}
for feature in iris.feature_names:
    renameDictionary[feature] = "".join(feature.split(" ")[0:2])
renameDictionary
# Converting the DataSet to Data Frame
iris_df = pd.DataFrame(iris['data'], columns = iris['feature_names'])
iris_df['target'] = pd.Series(iris['target'], name = 'target_values')
iris_df['target_name'] = iris_df['target'].replace([0,1,2],
[species for species in iris['target_names'].tolist()])
iris_df = iris_df.rename(columns = renameDictionary)
iris_df
```

```python
sns.scatterplot(data=iris_df, x="sepallength", y="petallength",hue="target_name",style="target_name")
```

#### Modelling

```python
X_train,X_test,y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.20)
from sklearn import linear_model
```

```python
model = linear_model.RidgeClassifier()
```

```python
np.logspace(-5,5)
```

```python
ridgeConstant = []
scoreArray = []
for x in np.logspace(-10,10,2000):
    model = linear_model.RidgeClassifier(alpha=x)
    model.fit(X_train,y_train)
    score = model.score(X_test,y_test)
    ridgeConstant.append(x)
    scoreArray.append(score)

sns.lineplot(x=ridgeConstant,y=scoreArray)
```

```python
# Best Value of Alpha
ridgeConstant[scoreArray.index(max(scoreArray))]
```

```python
model = linear_model.RidgeClassifier(alpha=ridgeConstant[scoreArray.index(max(scoreArray))])
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

##### Evaluation

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(cm,annot=True,)
plt.xlabel("Truth")
plt.ylabel("Predicted")
```
